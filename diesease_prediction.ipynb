```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('Disease Symptom Prediction.csv')
```

```python
data.head(15)
```

```python
data.describe()
```

```python
data.isna().sum()
```

```python
data.drop(columns=[i for i in data.columns[5:]],inplace=True)
```

```python
data.fillna(0,inplace=True)
```

```python
Symptom_1_map = data['Symptom_1'].value_counts().to_dict()
Symptom_2_map = data['Symptom_2'].value_counts().to_dict()
Symptom_3_map = data['Symptom_3'].value_counts().to_dict()
Symptom_4_map = data['Symptom_4'].value_counts().to_dict()
```

```python
data['Symptom_1'] = data['Symptom_1'].map(Symptom_1_map)
data['Symptom_2'] = data['Symptom_2'].map(Symptom_2_map)
data['Symptom_3'] = data['Symptom_3'].map(Symptom_3_map)
data['Symptom_4'] = data['Symptom_4'].map(Symptom_4_map)
```

```python
data.head(15)
```

```python
len(data['Disease'].unique())
```

```python
d = {}
for i,j in zip(data['Disease'].unique(),range(len(data['Disease'].unique()))):
  d[i] = j
```

```python
data['Disease'] = data['Disease'].map(d)
```

```python
data.head(15)
```

```python
import seaborn as sns
sns.countplot(data['Disease'])
```

```python
import tensorflow
from tensorflow import keras
```

```python
x = data.iloc[:,1:]
y = data.iloc[:,:1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=72,test_size=0.10,stratify=y)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scale_test = MinMaxScaler()
scale_train = MinMaxScaler()
```

```python
x_train_scaled = scale_train.fit_transform(x_train)
x_test_scaled = scale_test.fit_transform(x_test)
```

```python
y_train_cat = keras.utils.to_categorical(y_train,num_classes=41,dtype='float32')
y_test_cat = keras.utils.to_categorical(y_test,num_classes=41,dtype='float32')
```

```python
print(x_train.shape)
len(x_train)
```

```python
model = keras.Sequential([
                  keras.layers.Flatten(input_shape = (4,)),
                  keras.layers.Dense(3000,activation='elu'),
                  keras.layers.Dense(3000,activation='elu'),
                  keras.layers.Dense(2000,activation='elu'),
                  keras.layers.Dense(2000,activation='elu'),
                  keras.layers.Dense(1000,activation='elu'),
                  keras.layers.Dense(1000,activation='elu'),
                  keras.layers.Dense(500,activation='elu'),
                  keras.layers.Dense(500,activation='elu'),
                  keras.layers.Dense(41,activation='softmax')
])
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics='accuracy')
```

```python
model.fit(x_train_scaled,y_train,epochs=100,batch_size=64)
```

```python
model.evaluate(x_test_scaled,y_test)
```

```python
from sklearn.metrics import classification_report
```

```python
y_pred = np.round(model.predict(x_test_scaled))
y_real_pred = [[np.argmax(i)] for i in y_pred]
```

```python
print(classification_report(y_pred=y_real_pred,y_true=np.array(y_test)))
```
